```python
import cupy as cp

def get_arr(x, baseline_correct=False, baseline_size=1000):
    if type(x) is cp.ndarray:
        res = x.get()
    else:
        res = x
    
    if baseline_correct:
        res -= res[..., :baseline_size].mean(axis=-1, keepdims=True)

    return res
```

```python
frequencies = np.geomspace(5, 100, 80)
n_frequencies = len(frequencies)

n_samples = 3000

n_parcels = 200

cue_types = {'att' : 0, 'non': 1, 'res': 2}

n_types = len(cue_types)
```

```python
hit_subjectwise = np.zeros((len(subjects), n_frequencies, n_parcels, n_samples))
miss_subjectwise = np.zeros((len(subjects), n_frequencies, n_parcels, n_samples))

hit_amp_subjectwise = np.zeros((len(subjects), n_frequencies, n_parcels, n_samples))
miss_amp_subjectwise = np.zeros((len(subjects), n_frequencies, n_parcels, n_samples))

hit_evoked_subjectwise = np.zeros((len(subjects), n_frequencies, n_parcels, n_samples))
miss_evoked_subjectwise = np.zeros((len(subjects), n_frequencies, n_parcels, n_samples))

for subject_idx, subject_root in enumerate(subjects):
    # freqs  x parcels x trials x lags
    subject_files = glob.glob(os.path.join(subject_root, '*.npy'))
    
    subject_hit = np.zeros((n_frequencies, n_parcels, n_samples))
    subject_miss = np.zeros((n_frequencies, n_parcels, n_samples))
    
    subject_amp_hit = np.zeros((n_frequencies, n_parcels, n_samples))
    subject_amp_miss = np.zeros((n_frequencies, n_parcels, n_samples))
    
    subject_evoked_hit = np.zeros((n_frequencies, n_parcels, n_samples))
    subject_evoked_miss = np.zeros((n_frequencies, n_parcels, n_samples))
    
    subject_type_counter = 0

    for file_idx, fpath in enumerate(subject_files):
        subject_data = pickle.load(open(fpath, 'rb'))
        
        type_index = cue_types[subject_data['cue']]
        
        if type_index == 1: # consider only non paradigm! saves some memory
            subject_hit += get_arr(subject_data['hit']['pacf'], baseline_correct=False)
            subject_miss += get_arr(subject_data['miss']['pacf'], baseline_correct=False)
            
            subject_amp_hit += get_arr(subject_data['hit']['amplitude'], baseline_correct=False)
            subject_amp_miss += get_arr(subject_data['miss']['amplitude'], baseline_correct=False)
            
            subject_evoked_hit += get_arr(np.abs(subject_data['hit']['evoked']), baseline_correct=False)
            subject_evoked_miss += get_arr(np.abs(subject_data['miss']['evoked']), baseline_correct=False)
        
            subject_type_counter += 1    

    hit_subjectwise[subject_idx] = (subject_hit / subject_type_counter)
    miss_subjectwise[subject_idx] = (subject_miss / subject_type_counter)  
    
    hit_amp_subjectwise[subject_idx] = (subject_amp_hit / subject_type_counter)
    miss_amp_subjectwise[subject_idx] = (subject_amp_miss / subject_type_counter)  
    
    hit_evoked_subjectwise[subject_idx] = (subject_evoked_hit / subject_type_counter)
    miss_evoked_subjectwise[subject_idx] = (subject_evoked_miss / subject_type_counter)  
```

```python
cohort_pacf_diff = np.nanmean(hit_subjectwise - miss_subjectwise, axis=0)
cohort_amp_diff = np.nanmean(hit_amp_subjectwise, axis=(0)) - np.nanmean(miss_amp_subjectwise, axis=(0))
cohort_evoked_diff = np.nanmean(hit_evoked_subjectwise, axis=(0)) - np.nanmean(miss_evoked_subjectwise, axis=(0))
```

```python
def compute_surr_cohort_diff(x, y, n_rounds=100):
    xy = np.concatenate([x,y], axis=0)

    labels = np.array([0]*len(x) + [1]*len(y))

    res = np.zeros((n_rounds, *x.shape[1:]))

    for i in tqdm.tqdm(np.arange(n_rounds)):

        for j in range(x.shape[0]):
            if np.random.random() > 0.5:
                res[i] += x[j]
                res[i] -= y[j]
            else:
                res[i] -= x[j]
                res[i] += y[j]
        
        res[i] /= x.shape[0]

    return res

def jk_pacf_diff_sign(pacf_diff, surr_level_parcelwise):
    subject_mask = np.ones(pacf_diff.shape[0], dtype=bool)
    subject_mask[0] = False

    res_pos = np.zeros_like(pacf_diff)
    res_neg = np.zeros_like(pacf_diff)

    for i in range(pacf_diff.shape[0]):
        res_pos[i] = (pacf_diff[subject_mask].mean(axis=0) >= surr_level_parcelwise)
        res_neg[i] = (pacf_diff[subject_mask].mean(axis=0) <= -surr_level_parcelwise)
 

        subject_mask = np.roll(subject_mask, 1)

    return res_pos, res_neg
```

```python

if os.path.exists('tsdt_surr_level_parcelwise.npy'):
# if False:
    surr_level_parcelwise = np.load('tsdt_surr_level_parcelwise.npy')
else:
    surr_diff = compute_surr_cohort_diff(hit_subjectwise_finite, miss_subjectwise_finite, n_rounds=500)
    surr_level_parcelwise = np.zeros((1, 80, 200, 1))

    bar = tqdm.tqdm(total=200*80)
    for i in range(200):
        for j in range(80):
            surr_level_parcelwise[:, j, i] = np.percentile(surr_diff[:, j, i].max(axis=-1), 95)
            bar.update(1)

    del surr_diff
    np.save('tsdt_surr_level_parcelwise.npy', surr_level_parcelwise)
```
